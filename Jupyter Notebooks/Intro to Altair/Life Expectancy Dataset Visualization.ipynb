Visualizing the "countries" dataset with a regular line plot.

```python
import altair as alt
from vega_datasets import data

# Run this once per session, enables Altair rendering
alt.renderers.enable("notebook")

# Define dataframe as the countries dataset
df = data.countries()

# Display sample of data format
df.head()
```

```python

alt.Chart(df, height = 1000, width = 700).mark_line().encode(
    alt.X("year:N"),
    alt.Y("life_expect:Q", scale = alt.Scale(domain=[20,85])),
    alt.Color("country:N"),
)
```

Interactive version of chart that highlights closest line to cursor.
Adapted from Altair Gallery Example: https://altair-viz.github.io/gallery/multiline_highlight.html

```python
stocks = data.countries.url

highlight = alt.selection(type='single', on='mouseover',
                          fields=['country'], nearest=True)

base = alt.Chart(stocks, height = 1000, width = 700).encode(
    x='year:N',
    y='life_expect:Q',
    color=alt.Color('country:N', scale=alt.Scale(scheme='category20'))
)

points = base.mark_circle().encode(
    opacity=alt.value(0)
).properties(
    selection=highlight,
    width=600
)

lines = base.mark_line().encode(
    size=alt.condition(~highlight, alt.value(1), alt.value(3))
)

points + lines
```

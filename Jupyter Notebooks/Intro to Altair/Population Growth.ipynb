Visualizing population growth between years 1850-2000 in different age groups, spanning 5 years each, ages ranging from 0-90.

```python
import altair as alt
from vega_datasets import data

# Run this once per session, enables Altair rendering
alt.renderers.enable("notebook")

df = data.population()
df.head()
```

```python
alt.Chart(df).mark_line().encode(
    alt.X("year:N"),
    alt.Y("people:Q"),
    alt.Color("age")).interactive()
```

Streamgraph visualization

```python
alt.Chart(df).mark_area().encode(
    alt.X('year:N',
        axis=alt.Axis(format='%Y', domain=False, tickSize=0)
    ),
    alt.Y('sum(people):Q', stack='center'),
    alt.Color('age:N',
        scale=alt.Scale(scheme='category20b')
    )
).interactive()
```

It appears that each age group has been gradually increasing in size over the years, with ages 0-55 growing the most and ages 55-90 growing more slowly.

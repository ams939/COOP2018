Visualizing obesity rate in each U.S state, adaptation from Altair Choropleth Example code at: https://altair-viz.github.io/gallery/choropleth.html

```python
import altair as alt
from vega_datasets import data

# Run this once per session, enables Altair rendering
alt.renderers.enable("notebook")
```

```python
# Import data into dataframe
df = data.obesity()
df.head()
```

```python
states = alt.topo_feature(data.us_10m.url, 'states')

alt.Chart(states).mark_geoshape().encode(
    color='rate:Q'
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(df, 'id', ['rate'])
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300
)
```

It appears that Colorado is the state with the lowest obesity rate and Illinois is the state with the highest obesity rate.

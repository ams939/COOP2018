## Seattle Weather Exploration: Windspeed

This notebook is a part of the author's practise in getting familiar with Altair. The visualizations & the code for them has been inspired by Altair's "Seattle Weather Exploration" tutorial: https://altair-viz.github.io/case_studies/exploring-weather.html

All of the code here is based on the code found at the link above.

```python
import altair as alt
from vega_datasets import data

# Run this once per session, enables Altair rendering
alt.renderers.enable("notebook")

# Define dataframe as the seattle weather dataset
df = data.seattle_weather()
```

#### Visualization 1.
Visualizing windspeeds in records

```python
alt.Chart(df).mark_bar().encode(
    alt.X("wind", bin = True),
    alt.Y("count()")
)
```

As can be seen, windspeed does not have a normal distribution but appears to be slightly skewed to the right

#### Visualization 2.

Hypotheses: On days where there is a large temperature difference (large temperature range), daily windspeed recorded will also be greater.

Plotting:
Temperature range vs. wind speed

```python
# Define colors for markers on graph
scale = alt.Scale(domain=['sun', 'fog', 'drizzle', 'rain', 'snow'],
                  range=['#FFB013', '#808080', '#52A4FF', '#0066D8', '#490EE6'])

alt.Chart(df).mark_point().encode(
    alt.X("wind:Q", axis = alt.Axis(title = "Daily windspeed (mph)")),
    alt.Y("temp_range:Q", axis = alt.Axis(title = "Daily temperature range (C)")),
    alt.Color("weather", scale = scale),
).transform_calculate("temp_range", "datum.temp_max - datum.temp_min"
).properties(
    width=600,
    height=400
).interactive()
```

Conclusion: There does not appear to be a relationship between temperature range and windspeed for this data set

#### Visualization 3.

Hypotheses: On windy days, the daily average temperature will be lower

Plot average daily temperature vs. windspeed

```python
alt.Chart(df).mark_point().encode(
    alt.X("wind:Q", axis = alt.Axis(title = "Daily windspeed (mph)")),
    alt.Y("avg_temp:Q", axis = alt.Axis(title = "Average temperature (C)")),
    alt.Color("weather", scale = scale),
).transform_calculate("avg_temp", "(datum.temp_max + datum.temp_min) / 2"
).properties(
    width=600,
    height=400
).interactive()
```

It appears that there is no clear relationship between the average daily temperature and windspeed

#### Visualization 4.

Hypotheses: It is more windy in the winter months than the summer months

Plot mean wind vs. date (month)

```python
alt.Chart(df).mark_line().encode(
    alt.X("date:T", timeUnit="month"),
    alt.Y("average(wind)"),
)
```

It appears that the average wind in the summer months is lower than in the winter months.

#### Visualization 5.

Visualizing linkage between windspeed and weather type

```python
alt.Chart(df).mark_bar().encode(
    x = alt.X("wind", bin = True, axis = alt.Axis(title="Month of the Year")),
    y = "count()",
    color = alt.Color("weather", legend = alt.Legend(title = "Weather type"), scale = scale))
```

#### Visualization 6.

Visualizing the relationship between wind and precipitation

```python
alt.Chart(df).mark_point().encode(
    x = alt.X("wind:Q"),
    y = alt.Y("precipitation:Q"))
```

#### Visualization 7.

Visualizing how weather types are distributed over a period of time

```python
alt.Chart(df).mark_area().encode(
    alt.X("date:T", timeUnit = "month", axis = alt.Axis(domain=False, tickSize = 0)),
    alt.Y("count():Q", stack = "center", axis = None),
    alt.Color("weather", scale = scale)
).interactive()
```

```python
alt.Chart(df).mark_area().encode(
    alt.X("date:T", timeUnit = "year", axis = alt.Axis(domain=False, tickSize = 0)),
    alt.Y("count():Q", stack = "center", axis = None),
    alt.Color("weather", scale = scale)
).interactive()
```

Odd result, possibly incorrect visualization method.

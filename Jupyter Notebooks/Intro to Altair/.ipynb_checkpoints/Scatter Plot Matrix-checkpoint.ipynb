```python
import altair as alt
from vega_datasets import data

# Run this once per session, enables Altair rendering
alt.renderers.enable("notebook")

# Define dataframe as the countries dataset
df = data.iris()

# Display sample of data format
df.head()
```

```python
alt.Chart(df).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='species:N'
).properties(
    width=250,
    height=250
).repeat(
    row=['petalLength', 'petalWidth', 'sepalLength', 'sepalWidth'],
    column=['sepalWidth','sepalLength', 'petalWidth', 'petalLength']
).interactive()
```
